```python
import cv2
image = cv2.imread('s_futbol1.jpg')
B, G, R = cv2.split(image)
cv2.imshow("original", image)
cv2.waitKey(0)
cv2.imshow("blue", B)
cv2.waitKey(0)
cv2.imshow("Green", G)
cv2.waitKey(0)
cv2.imshow("red", R)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt

# Renkli bir görüntüyü yükle
image = cv2.imread('s_futbol1.jpg')

# BGR kanallarını ayır
b, g, r = cv2.split(image)

# Her bir kanalı gösterme
plt.subplot(1, 3, 1)
plt.imshow(b, cmap='gray')
plt.title('Mavi Kanal')

plt.subplot(1, 3, 2)
plt.imshow(g, cmap='gray')
plt.title('Yeşil Kanal')

plt.subplot(1, 3, 3)
plt.imshow(r, cmap='gray')
plt.title('Kırmızı Kanal')

plt.show()
```

```python
roi = image[100 : 500, 200 : 700]
cv2.imshow("ROI", roi)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
resize = cv2.resize(image, (500, 500))
cv2.imshow("Resized Image", resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
